```python
%%cython
cimport numpy
cimport cython
#from laplace_helper cimport L2_rel_error
#ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_jacobi_cython(numpy.ndarray[numpy.float_t, ndim=2] p, 
                            numpy.ndarray[numpy.float_t, ndim=1] y, 
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[numpy.float_t, ndim=2] pn
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        pn = p.copy()
        error = 0.0
        for j in range(1,ny-1):
            for i in range(1,nx-1):
                p[j,i] = .25 * (pn[j,i-1] + pn[j,i+1] + pn[j-1,i] + pn[j+1,i])
                #error += abs(p[j,i] - pn[j,i])
                
        for j in range(1,ny-1):
            p[j,-1] = .25 * (2*pn[j,-2] + pn[j+1,-1] + pn[j-1,-1])
        
        error = numpy.sqrt(numpy.sum((p - pn)**2)/numpy.sum(pn**2))
        iterations += 1
        
        #error = L2_rel_error(p, pn)
        
    return p, iterations
```

```python
t1 = time.time()
p = p0.copy()
eps = 1e-8
p, iterations = laplace2d_jacobi_cython(p, y, dx, dy, nx, ny, eps)
t2 = time.time()

print "Jacobi method took {:.4f} seconds over {} iterations at tolerance {}".\
        format(t2-t1, iterations, eps)
```

```python
%%cython
cimport numpy
cimport cython
ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_gauss_seidel_cython(numpy.ndarray[FTYPE_t, ndim=2] p, 
                            numpy.ndarray[FTYPE_t, ndim=1] y, 
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[FTYPE_t, ndim=2] pn
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        pn = p.copy()
        error = 0.0
        for j in range(1,ny-1):
            for i in range(1,nx-1):
                p[j,i] = .25 * (p[j,i-1] + p[j,i+1] + p[j-1,i] + p[j+1,i])
                                
        for j in range(ny):
            p[j,-1] = .25 * (2*p[j,-2] + p[j+1,-1] + p[j-1,-1])
        error = numpy.sqrt(numpy.sum((p - pn)**2)/numpy.sum(pn**2))
        iterations += 1
        
        
    return p, iterations
```

```python
%%cython
cimport numpy
cimport cython
ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_SOR_cython(numpy.ndarray[FTYPE_t, ndim=2] p, 
                            numpy.ndarray[FTYPE_t, ndim=1] y, 
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps, 
                            double omega):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[FTYPE_t, ndim=2] pn
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        pn = p.copy()
        error = 0.0
        for j in range(1,ny-1):
            for i in range(1,nx-1):
                p[j,i] = (1 - omega) * p[j,i] +\
                .25 * omega * (p[j,i-1] + p[j,i+1] + p[j-1,i] + p[j+1,i])
                error += abs(p[j,i] - pn[j,i])
                
        for j in range(ny):
            p[j,-1] = p[j,-2]
        iterations += 1
        #error /= float(nx*ny)
        
    return p, iterations
```

```python
%%cython
cimport numpy
cimport cython
ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_SRJ_cython(numpy.ndarray[FTYPE_t, ndim=2] p, 
                            numpy.ndarray[FTYPE_t, ndim=1] y, 
                            numpy.ndarray[FTYPE_t, ndim=1] wt,
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps, ):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[FTYPE_t, ndim=2] pn
    cdef double omega
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        for omega in wt:
            iterations += 1
            
            pn = p.copy()
            error = 0.0
            for j in range(1,ny-1):
                for i in range(1,nx-1):
                    p[j,i] = (1 - omega) * pn[j,i] +\
                    .25 * omega * (pn[j,i-1] + pn[j,i+1] + pn[j-1,i] + pn[j+1,i])
                    error += abs(p[j,i] - pn[j,i])

            for j in range(ny):
                p[j,-1] = p[j,-2]

            
                
            if error <= eps:
                return p, iterations
            #error /= float(nx*ny)
        
    return p, iterations
```
