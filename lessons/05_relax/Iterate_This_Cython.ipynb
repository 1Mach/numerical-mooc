```python
%%cython
cimport numpy
cimport cython
#ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_jacobi_cython(numpy.ndarray[numpy.float_t, ndim=2] p, 
                            numpy.ndarray[numpy.float_t, ndim=1] y, 
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[numpy.float_t, ndim=2] pn
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        pn = p.copy()
        error = 0.0
        for j in range(1,ny-1):
            for i in range(1,nx-1):
                p[j,i] = .25 * (pn[j,i-1] + pn[j,i+1] + pn[j-1,i] + pn[j+1,i])
                error += abs(p[j,i] - pn[j,i])
                
        for j in range(ny):
            p[j,-1] = p[j,-2]
        iterations += 1
        
    return p, iterations
```

```python
nx_vals = [32, 64, 128]

for nx in nx_vals:

    ny = nx

    L = 5.
    H = 5.

    x = numpy.linspace(0,L,nx)
    y = numpy.linspace(0,H,ny)

    dx = L/(nx-1)
    dy = H/(ny-1)

    p = numpy.zeros((ny, nx))

    p[:,0] = 0
    p[:,-1] = p[:,-2]
    p[0,:] = 0
    p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
    
    t1c = time.time()
    pjaccyth, itercyth = laplace2d_jacobi_cython(p, y, dx, dy, nx, ny, 1e-8)
    t2c = time.time()
    #print t2-t1
    
    p = numpy.zeros((ny, nx))

    p[:,0] = 0
    p[:,-1] = p[:,-2]
    p[0,:] = 0
    p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
    
    t1n = time.time()
    pjac, iternumpy = laplace2d_jacobi(p, y, dx, dy, 1e-8)
    t2n = time.time()
    #print t2-t1
    
    assert numpy.allclose(pjac, pjaccyth)
    
    print "Grid size: {}x{}    Numpy time: {} s for {} iterations".format(nx,nx, t2n-t1n, iternumpy)
    print "Grid size: {}x{}    Cython time: {} s for {} iterations".format(nx, nx, t2c-t1c, itercyth)
    
```

```python
%%cython
cimport numpy
cimport cython
ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_SOR_cython(numpy.ndarray[FTYPE_t, ndim=2] p, 
                            numpy.ndarray[FTYPE_t, ndim=1] y, 
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps, 
                            double omega):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[FTYPE_t, ndim=2] pn
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        pn = p.copy()
        error = 0.0
        for j in range(1,ny-1):
            for i in range(1,nx-1):
                p[j,i] = (1 - omega) * p[j,i] +\
                .25 * omega * (p[j,i-1] + p[j,i+1] + p[j-1,i] + p[j+1,i])
                error += abs(p[j,i] - pn[j,i])
                
        for j in range(ny):
            p[j,-1] = p[j,-2]
        iterations += 1
        #error /= float(nx*ny)
        
    return p, iterations
```

```python
%%cython
cimport numpy
cimport cython
ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_SRJ_cython(numpy.ndarray[FTYPE_t, ndim=2] p, 
                            numpy.ndarray[FTYPE_t, ndim=1] y, 
                            numpy.ndarray[FTYPE_t, ndim=1] wt,
                            double dx, 
                            double dy, 
                            int nx, 
                            int ny, 
                            double eps, ):
    cdef int iterations, i, j
    cdef double error
    cdef numpy.ndarray[FTYPE_t, ndim=2] pn
    cdef double omega
    
    iterations = 0
    error = 2*eps
    
    while error > eps:
        for omega in wt:
            iterations += 1
            
            pn = p.copy()
            error = 0.0
            for j in range(1,ny-1):
                for i in range(1,nx-1):
                    p[j,i] = (1 - omega) * pn[j,i] +\
                    .25 * omega * (pn[j,i-1] + pn[j,i+1] + pn[j-1,i] + pn[j+1,i])
                    error += abs(p[j,i] - pn[j,i])

            for j in range(ny):
                p[j,-1] = p[j,-2]

            
                
            if error <= eps:
                return p, iterations
            #error /= float(nx*ny)
        
    return p, iterations
```
