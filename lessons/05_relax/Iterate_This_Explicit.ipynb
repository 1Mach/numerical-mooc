Same Laplace problem:

\begin{equation}
  \begin{gathered}
p=0 \text{ at } x=0\\
\frac{\partial p}{\partial x}=0 \text{ at }x=1\\
p = 0 \text{ at }y = 0 \\
p = \sin \left(  \frac{\frac{3}{2}\pi x}{L} \right) \text{ at } y = 1
  \end{gathered}
\end{equation}

```python
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
```

```python
def L2_error(p, pn):
    return numpy.sqrt(numpy.sum((p - pn)**2))
```

```python
def plot2D(x, y, p):
    '''Creates 3D projection plot with appropriate limits and viewing angle
    
    Parameters:
    ----------
    x: array of float
        nodal coordinates in x
    y: array of float
        nodal coordinates in y
    p: 2D array of float
        calculated potential field
    
    '''
    fig = plt.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
    X,Y = numpy.meshgrid(x,y)
    surf = ax.plot_surface(X,Y,p[:], rstride=1, cstride=1, cmap=cm.coolwarm,
            linewidth=0, antialiased=False)

    ax.set_xlim(0,x[-1])
    ax.set_ylim(0,y[-1])
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.view_init(30,45)
```

```python
def laplace2d_jacobi(p, y, dx, dy, l2_target):
    '''Solves the diffusion equation with forward-time, centered scheme
    
    Parameters:
    ----------
    p: 2D array of float
        Initial potential distribution
    y: array of float
        Nodal coordinates in y
    dx: float
        Mesh size
    dy: float
        Mesh size
    l2_target: float
        Error target
        
    Returns:
    -------
    p: 2D array of float
        Potential distribution after relaxation
    '''
    
    l2norm = 1
    pn = numpy.empty_like(p)
    iteration_count = 0

    while l2norm > l2_target:
        pn = p.copy()
        p[1:-1,1:-1] = .25 * (pn[1:-1,2:]+pn[1:-1,:-2]+pn[2:,1:-1]+pn[:-2,1:-1])
            
        p[:,0] = 0
        p[0,:] = 0
        p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
        p[:,-1] = p[:,-2]
        
        l2norm = numpy.sum(numpy.abs(p - pn))
        
        iteration_count += 1
     
    return p, iteration_count
```

```python
%%cython
cimport numpy
cimport cython
ctypedef numpy.float_t FTYPE_t

import numpy

@cython.boundscheck(False)
def laplace2d_jacobi_cython():
    cdef int N, iterations, i, j, nx, ny
    cdef double dx, dy, error, l2_target, L, H
    cdef numpy.ndarray[FTYPE_t] y
    cdef numpy.ndarray[FTYPE_t, ndim=2] p, pn
    
    nx = 128
    ny = 128

    L = 5.
    H = 5.

    x = numpy.linspace(0,L,nx)
    y = numpy.linspace(0,H,ny)

    dx = L/(nx-1)
    dy = H/(ny-1)

    p = numpy.zeros((ny, nx))
    pn = numpy.zeros((ny, nx))
    p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
    
    iterations = 0
    l2_target = 1e-8
    error = 2*l2_target
    
    while iterations < 1e5 and error > l2_target:
        pn = p.copy()
        error = 0.0
        for j in range(1,ny-1):
            for i in range(1,nx-1):
                p[j,i] = .25 * (pn[j,i-1] + pn[j,i+1] + pn[j-1,i] + pn[j+1,i])
                error += abs(p[j,i] - pn[j,i])
                
        for j in range(ny):
            p[j,-1] = p[j,-2]
        iterations += 1
        error /= float(nx*ny)
        
    return iterations, p
```

```python
#@jit('f8(f8[:,:],f8,f8,f8,f8)')
@autojit
def laplace2d_jacobi_numba(p, y, dx, dy, nx, ny, l2_target):
    '''Solves the diffusion equation with forward-time, centered scheme
    
    Parameters:
    ----------
    p: 2D array of float
        Initial potential distribution
    y: array of float
        Nodal coordinates in y
    dx: float
        Mesh size
    dy: float
        Mesh size
    l2_target: float
        Error target
        
    Returns:
    -------
    p: 2D array of float
        Potential distribution after relaxation
    '''
    
    l2norm = 1
    pn = numpy.zeros(p.shape)
    iteration_count = 0
    
    while l2norm > l2_target:
        for j in xrange(1,ny-1):
            for i in xrange(1,nx-1):
                pn[j,i] = .25 * (p[j,i-1] + p[j,i+1] + p[j-1,i] + p[j+1,i])
                
        #for i in xrange(nx):
            #pn[0,i] = 0
            #pn[-1,i] = sin(1.5*pi*x[i]/x[-1])
        
        for j in xrange(ny):
            #pn[j,0] = 0
            pn[j,-1] = pn[j,-2]
                
        #p[1:-1,1:-1] = .25 * (pn[1:-1,2:]+pn[1:-1,:-2]+pn[2:,1:-1]+pn[:-2,1:-1])
            
        #p[:,0] = 0
        #p[0,:] = 0
        #p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
        #p[:,-1] = p[:,-2]
               
        l2norm = numbaL2(p, pn)
        
        for i in xrange(nx):
            for j in xrange(ny):
                p[j,i] = pn[j,i]
        
        iteration_count += 1
        
        #print "Iteration: {} -- Error: {}".format(iteration_count, l2norm)
     
    return p, iteration_count

#@jit('f8(f8[:,:],f8[:,:])')
@autojit
def numbaL2(p,pn):
    m, n = p.shape
    l2 = 0
    for i in xrange(m):
        for j in xrange(n):
            l2 += sqrt((p[i,j]-pn[i,j])**2)
    return l2

```

### SOR

$$p^{n+1}_{i,j} = (1 - \omega)p^n_{i,j} + \frac{\omega}{4} \left(p^n_{i,j-1} + p^n_{i,j+1} + p^n_{i-1,j} + p^n_{i+1,j} \right)$$

*  only convergent for $0 < \omega < 2$
*  faster than Gauss-Seidel only if $1 < \omega < 2$

$$\omega \approx \frac{2}{1+\frac{\pi}{nx}}$$

for a square lattice

cite: http://www.physics.buffalo.edu/phy410-505/2011/topic3/app1/index.html


```python
omega = 4./(2 + numpy.sqrt(4 + (numpy.cos(numpy.pi/nx)+numpy.cos(numpy.pi/ny))**2))
```

```python
def laplace2d_SOR(P, y, dx, dy, l2_target,omega):
    '''Solves the diffusion equation with forward-time, centered scheme
    
    Parameters:
    ----------
    p: 2D array of float
        Initial potential distribution
    y: array of float
        Nodal coordinates in y
    dx: float
        Mesh size
    dy: float
        Mesh size
    l2_target: float
        Error target
        
    Returns:
    -------
    p: 2D array of float
        Potential distribution after relaxation
    '''
    
    l2norm = 1
    #pn = numpy.empty_like(p)
    iteration_count = 0

    while l2norm > l2_target:
        pn = P.copy()
        P[1:-1,1:-1] = .25*(pn[1:-1,:-2]+pn[1:-1,2:]+pn[:-2,1:-1]+pn[2:,1:-1])
        
        
        
        P[1:-1] = (1-omega)*pn[1:-1] + omega*P[1:-1]
        
        P[:,0] = 0
        P[:,-1] = P[:,-2]
        P[0,:] = 0
        P[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
        
        l2norm = L2_error(P, pn)
        
        iteration_count += 1
     
    return P, iteration_count
```

```python
p = numpy.zeros((ny, nx))

p[:,0] = 0
p[:,-1] = p[:,-2]
p[0,:] = 0
p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])

pSOR, iterations = laplace2d_SOR(p, y, dx, dy, 1e-2, 1.1)
```

### Scheduled relaxed Jacobi

$$p^{n+1} = (D + \omega_i L)^{-1} (\omega_i b - [\omega_i U + (\omega_i -1)D]p^k)$$

From paper:
(Optimized for nx = ny = 64)

$\Omega = \{190.2, 0.9532\}$
$(q_1, q_2) = (1, 63)$

use $\omega_i$ for $q_i$ iterations, then increment i


P = 3 SRJ

```python
nx = 128
ny = 128

L = 5.
H = 5.

x = numpy.linspace(0,L,nx)
y = numpy.linspace(0,H,ny)

dx = L/(nx-1)
dy = H/(ny-1)

p = numpy.zeros((ny, nx))

p[:,0] = 0
p[:,-1] = p[:,-2]
p[0,:] = 0
p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])
```

```python
##P=2 N=128
omega =[425.8, 0.9742]
q = [1, 130]
```

```python
iterations = 0
L2 = 1
while L2 > 1e-8:
    for i,loopcount in enumerate(q):
        for j in range(loopcount):
            pn = p.copy()
            p[1:-1,1:-1] = (1 - omega[i])*pn[1:-1,1:-1] +\
                    .25*omega[i]*(pn[1:-1,:-2]+pn[1:-1,2:]+pn[:-2,1:-1]+pn[2:,1:-1])
            p[:,0] = 0
            p[:,-1] = p[:,-2]
            p[0,:] = 0
            p[-1,:] = numpy.sin(1.5*numpy.pi*x/x[-1])

            iterations += 1
        L2 = L2_error(p, pn)    
```
